### Load and Review Data

For this analysis we will use the: `data/ign_scores.csv`

- Use the `Platform` column as index

```python
file_path = '../data/ign_scores.csv'
ign_df = pd.read_csv(file_path, index_col='Platform')

# Print head
ign_df.head()
```

## Which Platform is the Best?

Create a `bar chart` that shows the average score for `racing` games for each platform.

```python
plt.figure(figsize=(16,6))
sns.barplot(x=ign_df['Racing'], y=ign_df.index)
plt.xlabel('Average Score')
```

## Task: Which one is the Best One for Racing Games?

If `Wii` platform is not the best, which platform is?

```python
platform_high_score = ign_df['Racing'].idxmax()
print(f'Platform with the highest avg score for racing games is {platform_high_score}')
```

### Task: All Possible Combinations

Create a heatmap of the average scores by genre in each cell

```python
plt.figure(figsize=(16,6))
plt.title('Heatmap')
sns.heatmap(data=ign_df, annot=True)
plt.xlabel('Genre')
```

## Task: Which Combination is Best?

- Which combination of genre and platform receives the highest average ratings?
- Which combination receives the lowest average rating?

```python
#highest_combo = ign_df['Genre'].idmax()
highest_combo = ign_df.stack().idxmax() # get tuple with columns
highest_combo_value = ign_df.stack().max()
lowest_combo = ign_df.stack().idxmin()
lowest_combo_value = ign_df.stack().min()

print(highest_combo, highest_combo_value, lowest_combo, lowest_combo_value)
```

## Scatter Plots

### Load the Dataset
We will use a (synthetic) datset of medical insurance charges: `data/insurance.csv`

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup complete')
```

```python
insurance_df = pd.read_csv('../data/insurance.csv')
insurance_df.head()
```

The plot suggests we have a positive correlation between BMI and charges.

- Higher BMI tends to pay more

To double check the strength of this relationship, add a `regression line` that best fits the data

```python
sns.regplot(x=insurance_df['bmi'], y=insurance_df['charges'])
```

### Color - Coded Scatter Plots

They're only good when you need to display the correlation between not two, but more variables.

We want to understand how `smokers` affects the relationship between `BMI` and `charges`

```python
# Scatter pplot of bmi vs charges AND smoker vs charges
# x = bmi, y = charges, hue = smoker
sns.scatterplot(x=insurance_df['bmi'], y=insurance_df['charges'], hue=insurance_df['smoker'])
```

Finally, we can adapt the design  of a scatterplot to feature a `categorical variable` (like 'smoker') on one of the main axis. Some ppl call this plot **Categorical Scatter Plot**

```python
sns.swarmplot(data=insurance_df, x=insurance_df['smoker'], y=insurance_df['charges'])
```
