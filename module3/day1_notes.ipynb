### Identify Missing Values

````python
# find the number of missing values
df_data.isnull()
````

````python
# missing values by percentage
#(df_data.isnull().sum/len)
````

````python
# remove the company column/feature
df_data.drop('company', inplace=True, axis=1) # axis 1 is columns
````

````python
df_data.head()
````

````python
# fill children feature with zeros
df_data.children = df_data.children.fillna(0)
````

````python
# missing values by percentage
(df_data.isnulll().sum()/len(df_data))*100
````

````python
# unique values in the country feature
df_data['country'].value_counts()
````

````python
# drop all rows with missing data values in the country feature
df_data = df_data.dropna(subset=['country']) # dropna = drop not available
````

## Duplicates

Practice how to handle duplicate values

````python
# Total number of duplicates in the dataset
df_data.duplicated().sum()
````

````python
# Extract the dataframe containing the duplicated valuse
df_data.loc[df_data.duplicated(), :] # : = all
````

````python
# drop the duplicate records
df_data.drop_duplicates(inplace=True, keep='first') #keep the first, drop the rest. can also do keep last
````

````python
# how to know the change? print the shape of the dataframe
df_data.shape
````

## Different Datatypes

### Datetime


````python
# convert the column "object" to a given type, in this case, datetime
df_data['reservation_status_date'] = pd.to_datetime(df_data['reservation_status_date'])
````

````python
# we just overwrote a column. verify.
df_data.info()
````

````python
# get the year, month, and day
df_data['reservation_status_year'] = df_data['reservation_status.date'].dt.year
df_data['reservation_status_month'] = df_data['reservation_status.date'].dt.month
df_data['reservation_status_day'] = df_data['reservation_status.date'].dt.day

df_data.head()
````

````python
# now we can drop the temporary column datetime
df_data.drop('reservation_status_date', inplace=True, axis=1)
````

## Bad or Illogical Records or values


````python
# Check how many records have 0 for 'adult'
(df_data['adults'] == 0).sum()

# Drop any records whose 'adult' column is 0
# use boolean indexes to filter/get your records
df_data.drop(df_data[df_data['adults'] == 0].index, inplace=True)
````
