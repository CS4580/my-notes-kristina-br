# Data Correlation

 - Columns that are redundant
 - Change of value in one column yield an almost constant change in another
 - Pair have either high (close to 1) or low (closer to 0) correlation scores

```python
# since version 2.0 of Pandas, use numeric_only=True
df_corr = df_yeast.corr(numeric_only=True)
print(df_corr)
```
